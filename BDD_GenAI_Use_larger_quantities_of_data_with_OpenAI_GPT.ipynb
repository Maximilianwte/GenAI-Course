```python
!git clone https://github.com/Maximilianwte/GenAI-Course.git
!pip install openai --q
import pandas as pd
pd.set_option('display.max_colwidth', None)
import os
import numpy as np
os.environ["OPENAI_API_KEY"] = "INPUT_OPENAI_TOKEN_HERE"

from openai import OpenAI
client = OpenAI()
```

```python
df = pd.read_csv('/content/GenAI-Course/Datasets/vision_pro_reviews.csv')
df.head(3)
```

```python
df_reviews = df.loc[df.contains_review == 1, ['id', 'translated_text']].sample(frac=1)
df_reviews
```

```python
def predict_gpt(texts):
      messages = [
          {"role": "user", "content": f"""What is written in the text?
          text: {texts}"""},
      ]
      completion = client.chat.completions.create(
          model='gpt-4o-mini',
          messages=messages,
          seed=1,
          max_tokens=200,
          temperature=0.5
      )
      return completion.choices[0].message.content.lower()
```

```python
selected_reviews = df_reviews.loc[:, 'translated_text'].values[0:2]
selected_reviews
```

```python
response = predict_gpt(selected_reviews)
print(response)
```
